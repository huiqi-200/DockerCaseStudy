```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
```

```python
df = pd.read_csv(f"Singapore_Covid_Data\singapore_covid19_data.csv")
```

```python


# Ensure the date column is in datetime format
df['Date'] = pd.to_datetime(df['Date'])

# # Group by the date column and count the number of rows for each date
# date_counts = df['Date'].value_counts().sort_index().reset_index()
# date_counts.columns = ['Date', 'Active']

# Plot the counts over the date column using Seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Date', y='Active')
plt.xlabel('Date')
plt.ylabel('Count of Rows')
plt.title('Count of Rows Over Date')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Create an interactive plot with a slider bar using Plotly
fig = px.line(df, x='Date', y='Active', title='Count of Rows Over Date')
fig.update_xaxes(rangeslider_visible=True)
fig.update_layout(xaxis_title='Date', yaxis_title='Count of Rows')
```
